```python
# now lets check whetrer the adr of reservation canceled is high or adr of reservation not cancelled
```

```python
cancelled_adr = cancel_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_adr.reset_index(inplace=True)
cancelled_adr.sort_values('reservation_status_date' , inplace=True)
cancelled_adr
```

```python
not_cancel_data = df[df['is_canceled'] == 0]
not_cancelled_adr = not_cancel_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_adr.reset_index(inplace=True)
not_cancelled_adr.sort_values('reservation_status_date' , inplace=True)
not_cancelled_adr
```

```python
plt.figure(figsize=(15,8))
plt.title('Average daily rate')
plt.plot(not_cancelled_adr['reservation_status_date'],not_cancelled_adr['adr'] , label = 'not canceled')
plt.plot(cancelled_adr['reservation_status_date'],cancelled_adr['adr'] , label = ' canceled')
plt.legend()
plt.show()
```

 Cancellation rate rises as the price rises, hotel can work on their pricing strategies, and reducing their price on weekends, they can also conduct campaigns, and also work on quality of the room
